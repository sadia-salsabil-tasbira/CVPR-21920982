```python
# Selection Sort

A = [16, 89, 7, 75, 66, 53, 44, 23, 34, 97]
  
print("Unsorted Array: ", A)
print("\n Sorting...")

for i in range(len(A)):  
    min_idx = i
    for j in range(i+1, len(A)):
        if A[min_idx] > A[j]:
            min_idx = j
    
    print("Minimum value:", A[min_idx], "... Swapping index:", i, "<->", min_idx, " & value:",  A[i], "<->", A[min_idx],)
    print("Current array", A, "\n")
    # Swapping 
    A[i], A[min_idx] = A[min_idx], A[i]
    
print ("\n Sorted array", A)
```

```python
# Quick sort
 
 
def partition(A, low, high):
    i = (low-1)         
    pivot = A[high]     
    print("Pivot: ", pivot)
    for j in range(low, high):
        print("Comparing ", A[j], "&", pivot)
        if A[j] <= pivot:
            i = i+1
            A[i], A[j] = A[j], A[i]
 
    A[i+1], A[high] = A[high], A[i+1]
    print("After putting pivot in the right position:", A, "\n")
    return (i+1)
 
 
def quickSort(A, low, high):
    if len(A) == 1:
        return A
    if low < high:
        pi = partition(A, low, high)
        
        quickSort(A, low, pi-1)
        quickSort(A, pi+1, high)
 
 

A = [16, 89, 7, 75, 66, 53, 44, 23, 34, 97]
n = len(A)
quickSort(A, 0, n-1)

print("Sorted array:", A)
```
