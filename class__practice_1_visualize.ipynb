```python
import matplotlib.pyplot as plt
import random
import math
from IPython import display
```

```python
X1 = [0,0,1,1]
X2 = [0,1,0,1]
Y = [0,0,0,1] #and
#Y = [0,1,1,1] #or
```

```python
plt.figure()

for i in range(len(Y)):
    if Y[i] == 0:
        plt.plot(X1[i], X2[i], "ro")
    else:
        plt.plot(X1[i], X2[i], "go")
        
plt.show()
```

```python
W = [random.random(), random.random()]
b = random.random()

print(W, b)
```

```python
lr=0.01
theta=0.5
```

```python
epoch = 0
count = 0
loss = []

while count < 4:    
    total_error = 0
    for i in range(len(Y)):
        weighted_sum = W[0]*X1[i] + W[1]*X2[i] + b
        pred = 0
        if weighted_sum > theta:
            pred = 1
        error =  Y[i] - pred
        
        if error != 0:
            count = 0
            total_error += math.sqrt(error**2)
            W[0] += error * X1[i] * lr
            W[1] += error * X2[i] *lr
            b    += error * lr
        else:
            count += 1
    
    loss.append(1/4*total_error)
    
    plt.figure(figsize=(20,5))
    display.clear_output(wait=True)
    plt.subplot(1,2,1)
    
    for i in range(len(Y)):
        if Y[i] == 0:
            plt.plot(X1[i], X2[i], "ro")
        else:
            plt.plot(X1[i], X2[i], "go")
        

    xintr=(theta-b)/W[0]
    yintr=(theta-b)/W[1]
    slope=-yintr/xintr
    xx=plt.xlim()
    yy = []
    for x in xx:
        yy.append(slope*x+yintr)
    plt.plot(xx,yy)
    plt.subplot(1,2,2)
    plt.plot(loss)
    plt.show()

    epoch += 1
```
